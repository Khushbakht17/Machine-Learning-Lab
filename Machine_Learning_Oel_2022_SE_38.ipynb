## Data loading

### Subtask:
Load the provided CSV files into pandas DataFrames.


**Reasoning**:
Import the pandas library and load the two CSV files into pandas DataFrames.



```python
import pandas as pd

try:
    train_df = pd.read_csv('mnist_train.csv')
    test_df = pd.read_csv('mnist_test.csv')
    display(train_df.head())
    display(test_df.head())
except FileNotFoundError:
    print("Error: One or both of the CSV files were not found.")
    train_df = None
    test_df = None
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    train_df = None
    test_df = None
```

## Data exploration

### Subtask:
Explore the loaded datasets (`train_df` and `test_df`) to understand their basic characteristics.


**Reasoning**:
Explore the datasets by examining their shapes, target variable distributions, missing values, and descriptive statistics. Generate plots for target variable distributions.



```python
import matplotlib.pyplot as plt

# Data Shape
print("Train Data Shape:", train_df.shape)
print("Test Data Shape:", test_df.shape)

# Target Variable Distribution
print("\nTrain Data - Target Variable Distribution:")
print(train_df['label'].value_counts())
plt.figure(figsize=(10, 5))
train_df['label'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Digits in Train Data')
plt.xlabel('Digit')
plt.ylabel('Frequency')
plt.show()

print("\nTest Data - Target Variable Distribution:")
print(test_df['label'].value_counts())
plt.figure(figsize=(10, 5))
test_df['label'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Distribution of Digits in Test Data')
plt.xlabel('Digit')
plt.ylabel('Frequency')
plt.show()

# Missing Values
print("\nTrain Data - Missing Values:")
print(train_df.isnull().sum())
print("\nTest Data - Missing Values:")
print(test_df.isnull().sum())

# Descriptive Statistics
print("\nTrain Data - Descriptive Statistics for Pixel Values:")
display(train_df.drop('label', axis=1).describe())
print("\nTest Data - Descriptive Statistics for Pixel Values:")
display(test_df.drop('label', axis=1).describe())
```

## Data preparation

### Subtask:
Prepare the data for model training by separating features and target variables, and normalizing pixel values.


**Reasoning**:
Separate features and target variables, normalize pixel values, and check/convert data types.



```python
# Feature/Target Separation
X_train = train_df.drop('label', axis=1)
y_train = train_df['label']
X_test = test_df.drop('label', axis=1)
y_test = test_df['label']

# Normalization
X_train = X_train / 255.0
X_test = X_test / 255.0

# Data Type Check and Conversion (optional but recommended)
if not pd.api.types.is_numeric_dtype(X_train.values.dtype):
    X_train = X_train.astype('float64')
if not pd.api.types.is_numeric_dtype(X_test.values.dtype):
    X_test = X_test.astype('float64')

print(X_train.dtypes)
print(X_test.dtypes)
```

## Data splitting

### Subtask:
Split the training data (`X_train`, `y_train`) into training and validation sets.


**Reasoning**:
Split the training data into training and validation sets using train_test_split and print the shapes of the resulting datasets.



```python
from sklearn.model_selection import train_test_split

X_train_split, X_val, y_train_split, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)

print("X_train_split shape:", X_train_split.shape)
print("X_val shape:", X_val.shape)
print("y_train_split shape:", y_train_split.shape)
print("y_val shape:", y_val.shape)
```

## Model training

### Subtask:
Train several classification models on the prepared MNIST dataset.


**Reasoning**:
Import necessary libraries and train the specified classification models using the provided training data.



```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

# Initialize the models with default hyperparameters
logreg_model = LogisticRegression(max_iter=1000)  # Increased max_iter to ensure convergence
rf_model = RandomForestClassifier()
svc_model = SVC()
mlp_model = MLPClassifier(max_iter=500) # Reduced max_iter for faster execution

# Train the models
models = {}

try:
    logreg_model.fit(X_train_split, y_train_split)
    models['LogisticRegression'] = logreg_model
    print("Logistic Regression trained successfully.")
except Exception as e:
    print(f"Error training Logistic Regression: {e}")

try:
    rf_model.fit(X_train_split, y_train_split)
    models['RandomForestClassifier'] = rf_model
    print("Random Forest trained successfully.")
except Exception as e:
    print(f"Error training Random Forest: {e}")

try:
    svc_model.fit(X_train_split, y_train_split)
    models['SVC'] = svc_model
    print("SVC trained successfully.")
except Exception as e:
    print(f"Error training SVC: {e}")

try:
    mlp_model.fit(X_train_split, y_train_split)
    models['MLPClassifier'] = mlp_model
    print("MLP trained successfully.")
except Exception as e:
    print(f"Error training MLP: {e}")
```

## Model optimization

### Subtask:
Optimize the hyperparameters of the trained classification models using the validation set.


**Reasoning**:
Optimize the hyperparameters of the trained classification models using the validation set.



```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Define the parameter grid for each model
param_grids = {
    'LogisticRegression': {'C': [0.1, 1, 10], 'penalty': ['l1', 'l2']},
    'RandomForestClassifier': {'n_estimators': [50, 100], 'max_depth': [None, 10], 'min_samples_split': [2, 5]},
    'SVC': {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf'], 'gamma': [0.001, 0.01]},
    'MLPClassifier': {'hidden_layer_sizes': [(50,), (100,)], 'activation': ['relu', 'tanh'], 'solver': ['adam'], 'alpha': [0.0001, 0.001]}
}

best_models = {}
best_scores = {}

for model_name, model in models.items():
    if model_name in param_grids:
        grid_search = GridSearchCV(model, param_grids[model_name], cv=3, scoring='accuracy', n_jobs=-1)
        try:
            grid_search.fit(X_val, y_val)
            best_models[model_name] = grid_search.best_estimator_
            best_scores[model_name] = grid_search.best_score_
            print(f"Best hyperparameters for {model_name}: {grid_search.best_params_}")
            print(f"Best score for {model_name}: {grid_search.best_score_}")
        except Exception as e:
            print(f"Error tuning {model_name}: {e}")
    else:
        print(f"No parameter grid defined for {model_name}")
```

## Model evaluation

### Subtask:
Evaluate the performance of the optimized models on the test set (`X_test`, `y_test`)


**Reasoning**:
Evaluate the performance of the optimized models on the test set (X_test, y_test) using the specified metrics.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import pandas as pd

evaluation_results = {}

for model_name, model in best_models.items():
    try:
        y_pred = model.predict(X_test)

        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='weighted')
        recall = recall_score(y_test, y_pred, average='weighted')
        f1 = f1_score(y_test, y_pred, average='weighted')
        conf_matrix = confusion_matrix(y_test, y_pred)

        evaluation_results[model_name] = {
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1_score': f1,
            'confusion_matrix': conf_matrix
        }

        print(f"Evaluation results for {model_name}:")
        print(f"  Accuracy: {accuracy}")
        print(f"  Precision: {precision}")
        print(f"  Recall: {recall}")
        print(f"  F1-score: {f1}")
        print(f"  Confusion Matrix:\n{conf_matrix}")

    except Exception as e:
        print(f"Error evaluating {model_name}: {e}")

results_df = pd.DataFrame(evaluation_results).T
display(results_df)
```

## Data visualization

### Subtask:
Visualize the performance metrics of the trained models.


**Reasoning**:
Visualize the performance metrics and confusion matrices of the trained models using bar charts and heatmaps.



```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Plotting performance metrics
metrics = ['accuracy', 'precision', 'recall', 'f1_score']
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.ravel()
for i, metric in enumerate(metrics):
    results_df[metric].plot(kind='bar', ax=axes[i], color=['skyblue', 'lightcoral', 'lightgreen', 'orange'])
    axes[i].set_title(f'{metric.capitalize()} Comparison')
    axes[i].set_ylabel(metric.capitalize())
    axes[i].set_xticklabels(results_df.index, rotation=45, ha='right')
plt.tight_layout()
plt.savefig('performance_metrics.png')
plt.show()


# Plotting confusion matrices
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.ravel()
for i, (model_name, row) in enumerate(results_df.iterrows()):
    conf_matrix = row['confusion_matrix']
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', ax=axes[i])
    axes[i].set_title(f'Confusion Matrix for {model_name}')
    axes[i].set_xlabel('Predicted Label')
    axes[i].set_ylabel('True Label')
plt.tight_layout()
plt.savefig('confusion_matrices.png')
plt.show()
```
